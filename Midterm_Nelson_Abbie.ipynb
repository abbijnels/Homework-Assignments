```python
input = [2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3]

import matplotlib.pyplot as plt
import numpy as np

frequency = {}
for item in input:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1

sorted_frequency = sorted(frequency.items())   
for i in sorted_frequency:
    print(i[0],':', i[1])
    
print(' ')

sorted_frequency = sorted(frequency.items())   
for k in sorted_frequency:
    print('Here', k[0], 'occurs', k[1], 'times')

plt.bar(list(frequency.keys()), frequency.values(), color='g')
plt.xlabel("List Item")
plt.ylabel("Frequency")
plt.title('Unsorted List Frequency Results')
plt.show()

import json

with open('frequency.json', 'w') as fp:
    json.dump(sorted_frequency, fp)
```

```python
import csv
import pandas as pd

file = open('Amazon.csv')
csvreader = csv.reader(file)

data = pd.read_csv('Amazon.csv')
data = data.fillna(0)
data

# total amount spent on Amazon
total_spent = data['Item Total'].sum()
print('Total amount spent on Amazon:', total_spent)

# mean value across all purchased items
avg_price = data['Item Total'].mean()
print('Average Value of Items:', avg_price)

# median value across all purchased items
median = data['Item Total'].median()
print('Median Value of Items:', median)

# Smallest value across all purchased items
min = data['Item Total'].min()
print('Smallest Value of Items:', min)

# Largest value across all purchased items
max = data['Item Total'].max()
print('Largest Value of Items:', max)


# Sort date format 
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Group purchases by date
daily_orders = data.groupby('Order Date').sum()['Item Total']

# Plot bar graph of purchases by date 
daily_orders.plot.bar(figsize=(35,10), color='purple')


plt.xlabel("Order Date")
plt.ylabel("Amount Spent")
plt.title('Amazon Spending History')
plt.show()
```
