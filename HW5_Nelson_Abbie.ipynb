```python
import matplotlib.pyplot as plt
import numpy as np 

# plot 1:
plt.subplot(1, 4, 1)

x1 = ([0, 0, 5, 5, 7, 10, 10, 12, 10, 20])
y2 = ([112, 102, 93, 99, 86, 87, 88, 85, 80, 85])

plt.scatter(x1, y2, color = 'blue')
plt.title('title 1')


# plot 2:
plt.subplot(1, 4, 2)

x2 = np.array([0, 1, 2, 3])
y2 = np.array([3, 8, 1, 10])

x3 = np.array([0, 1, 2, 5 ])
y3 = np.array([6, 4, 2, 11])

plt.plot(x2, y2,'blue')
plt.plot(x3, y3,'orange')
plt.title('title 2')


# plot 3:
plt.subplot(1, 4, 3)

X = ['A', 'B', 'C', 'D']
Y = [3, 5, 8, 10]

plt.bar(X, Y, color = 'blue')
plt.title('title 3')


# plot 4:
plt.subplot(1, 4, 4)

x = np.array([0, 85, 87.5, 95, 100, 110, 112.5, 118, 120, 125])
y = np.array([240, 250, 260, 270, 280, 295, 305, 310, 325, 340])

plt.plot(x, y, color = 'blue', marker = '*')
plt.title('title 4')

plt.subplots_adjust(wspace=0.6)
plt.show()
```
