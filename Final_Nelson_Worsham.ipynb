```python
import turtle


def draw_rectangle(length, height, color, x = 0, y = 0):
    
    # tells curser speed to fill and to fill color based on r,g,b parameter
    t = turtle.Turtle()
    t.speed(1000)
    t.fillcolor(color)
    t.hideturtle()
    
    # tells curser where to start drawing the rectangle
    t.up()
    t.goto(x,y)
    t.down()
    t.begin_fill()
    
    # given l
    for i in range(2):
        t.forward(length)
        t.right(90)
        t.forward(height)
        t.right(90)
    t.end_fill()
    t.setx(length)
    
# when called, get_color returns the color-specific integers to the draw_flag func
def get_color(color):
    if color == "red":
        return 1,0,0
    if color == "white":
        return 1,1,1
    if color == "blue":
        return 0,0,1
    else:
        return 0,0,0
    
def draw_star(size, color, x = 0, y = 0):
    # tells curser speed to fill and to fill color based on r,g,b parameter
    t = turtle.Turtle()
    t.speed(1000)
    t.pencolor(get_color("white"))
    t.up()
    turn = 144
    t.goto(x,y)
    t.fillcolor(color)
    t.hideturtle()
    t.down()
    
    # draw and fill the stars 
    t.begin_fill()
    for i in range(5):
        t.forward(size)
        t.right(turn)

    t.end_fill()
        
    
def draw_flag(height):
    
    # set flag dimensions proportional American flag dimensions
    
    # proportional flag length
    flag_length = 1.9 * float(height)
    
    # proportional flag height
    row_height = height * (7/13)
    
    # proportional length of blue star region
    stars_box = flag_length * (2/5)
    
    # row height 
    stripe = height / 13
    
    # row height for stars
    star = stripe * (4/5)
    
    # creates the number of rows in blue star region
    star_row = row_height / 10
    
    # creates the number of columns in blue star region
    star_column = stars_box / 14
    
    # draw the red rectangle for flag base
    color = get_color("red")
    draw_rectangle(flag_length, height, color)
   
    # draw the blue rectangle over red for the blue star area
    color = get_color("blue")
    draw_rectangle(stars_box, row_height, color)

    # draw rows to the right of the blue star area
    for white_row_number_top in range(1, 8, 2):
        color = get_color("white")
        draw_rectangle(flag_length - stars_box, stripe, color, stars_box, 
        -stripe * white_row_number_top)

    for red_row_number_top in range(2, 8, 2):
        color = get_color("red")
        draw_rectangle(flag_length - stars_box, stripe, color, stars_box,
        -stripe * red_row_number_top)

    
    # draw rows below blue star area; row spans entire length of flag
    for white_row_number_bottom in range(0, 6, 2):
        x = 0
        color = get_color("white")    
        draw_rectangle(flag_length, stripe, color, x, -row_height + 
        (-stripe * white_row_number_bottom))

    for red_row_number_bottom in range(1, 7, 2):
        x = 0
        color = get_color("red")    
        draw_rectangle(flag_length, stripe, color, x, -row_height + 
        (-stripe * red_row_number_bottom))
    
    
# draw the stars in the blue region    
    color = get_color("white")

# draw 30 stars in rows of 5 by 6:
    for column1 in range(1,12,2):
        for row1 in range(1,10,2):
            draw_star(star, color, star_row * column1 , -star_column * row1 )

# draw remaining 20 stars in between in rows of 5 by 4     
    for column2 in range(2,12,2):
        for row2 in range(2,10,2):
            draw_star(star,color, star_row * column2, -star_column * row2)

def main():
    draw_flag(200)

if __name__ == "__main__":
    main()    
    
    
    
    
```
